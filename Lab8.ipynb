## Part A: Starting the Spark Environment

```python
import sys
import os

# Set SPARK_HOME
os.environ["SPARK_HOME"] = "/opt/spark"

# Add Spark libraries to Python path
sys.path.append("/opt/spark/python")
sys.path.append("/opt/spark/python/lib/py4j-0.10.9.7-src.zip")

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("SparkLab") \
    .master("local[*]") \
    .getOrCreate()

sc = spark.sparkContext

print("Spark started successfully")
```

## Part B: Creating and Working with RDDs

RDD (Resilient Distributed Dataset) is the fundamental data structure in Apache Spark.
RDDs allow distributed processing of data across multiple nodes.

In this section, we will create an RDD from a Python list and perform basic operations.

```python
data = [5, 10, 15, 20, 25]

rdd = sc.parallelize(data)

print("RDD elements:", rdd.collect())
```

### Task 1

Apply a **map transformation** to multiply each element of the RDD by 3.

```python
result = rdd.map(lambda x: x * 3)

result.collect()
```

### Task 2

Filter the numbers that are greater than 15.

```python
filtered = rdd.filter(lambda x: x > 15)

filtered.collect()
```

### Task 3

Compute the sum of all elements in the RDD using a reduce operation.

```python
total = rdd.reduce(lambda a, b: a + b)

print("Sum of elements:", total)
```

## Part C: Word Count using Apache Spark

In the previous lab, you implemented the **Word Count problem using the MapReduce model**.

In this section, you will perform the same task using **Apache Spark**.

Spark simplifies distributed data processing by allowing transformations and actions directly on distributed datasets.

### Step 1: Load the Text Dataset

```python
text = sc.textFile("/home/student/spark_lab8_dataset.txt")

text.take(5)
```

### Step 2: Split Lines into Words

We use the **flatMap transformation** to split each line into words.

```python
words = text.flatMap(lambda line: line.split())

words.take(10)
```

### Step 3: Map Each Word to a Count

```python
word_pairs = words.map(lambda word: (word, 1))

word_pairs.take(10)
```

### Step 4: Reduce by Key

Aggregate counts for each word

```python
word_counts = word_pairs.reduceByKey(lambda a, b: a + b)

word_counts.take(10)
```

### Task 4

Display the **10 most frequent words** in the dataset.

```python
top_words = word_counts.takeOrdered(10, key=lambda x: -x[1])

top_words
```

### Task 5

Find the **total number of unique words** in the dataset.

```python
unique_words = word_counts.count()

print("Total unique words:", unique_words)
```

## Part D: Spark vs MapReduce Comparison

In the previous lab, you implemented data processing using the MapReduce model.  
Apache Spark provides a more efficient framework for performing similar tasks.

In this section, you will observe how Spark simplifies distributed data processing compared to traditional MapReduce.

### Conceptual Comparison

MapReduce performs computations in two major stages:

1. Map Phase  
2. Reduce Phase  

Apache Spark, on the other hand, allows multiple transformations to be chained together before performing actions, making the process faster and more flexible.

### Task 6

Perform a simple Spark transformation pipeline using the log dataset.

Your goal is to determine **how many times each webpage was accessed**.

```python
logs = sc.textFile("/home/student/server_logs.txt")

pages = logs.map(lambda line: line.split('"')[1].split()[1])

pages.take(10)
```

```python
page_pairs = pages.map(lambda page: (page,1))

page_pairs.take(10)
```

```python
page_counts = page_pairs.reduceByKey(lambda a,b: a+b)

page_counts.collect()
```
